```python
import plotly.express as px

def plot_unemployment_rates(df):
    try:
        fig = px.choropleth(
            df,
            locations='location',
            locationmode='country names',
            color='Unemployment_Rate',
            title='Unemployment Rate by Country',
            color_continuous_scale='viridis',
            range_color=[0, 15],
            labels={'Unemployment_Rate': 'Unemployment Rate'}
        )

        fig.update_geos(showocean=True, oceancolor="LightBlue", showland=True, landcolor="LightGreen")
        fig.show()

    except Exception as e:
        print(f"An error occurred: {e}")
```

# ***Khả năng đọc thành thạo, trình độ Toán theo quốc gia***

---



```python
def plot_proficiency_levels(df):
    try:
      # Cột biểu thị mức độ thành thạo môn đọc và toán
        proficiency_columns = [
            'Grade_2_3_Proficiency_Reading', 'Grade_2_3_Proficiency_Math',
            'Primary_End_Proficiency_Reading', 'Primary_End_Proficiency_Math',
            'Lower_Secondary_End_Proficiency_Reading', 'Lower_Secondary_End_Proficiency_Math'
        ]

       # Trích xuất các cột tồn tại trong tập dữ liệu
        available_columns = [col for col in proficiency_columns if col in df.columns]

       # Tính mức độ thành thạo trung bình về đọc và toán ở tất cả các cấp học
        average_values = [df[col].mean() for col in available_columns]

        if len(average_values) >= 2:  # Kiểm tra xem có giá trị nào cần vẽ cho cả môn đọc và toán không
            labels = ['Grade 2-3', 'Primary', 'Lower Secondary']
            reading_values = average_values[::2] # Trích xuất các giá trị trình độ đọc
            math_values = average_values[1::2]   # Trích xuất các giá trị trình độ toán học

            x = np.arange(len(labels)) # vị trí nhãn

            plt.figure(figsize=(10, 6))

            # Biểu đồ thanh về trình độ đọc
            plt.subplot(1, 2, 1)
            plt.bar(x, reading_values, color='skyblue')
            plt.xlabel('Education Levels')
            plt.ylabel('Proficiency Level')
            plt.title('Average Reading Proficiency')
            plt.xticks(x, labels)

           # Biểu đồ thanh về Trình độ Toán học
            plt.subplot(1, 2, 2)
            plt.bar(x, math_values, color='salmon')
            plt.xlabel('Education Levels')
            plt.ylabel('Proficiency Level')
            plt.title('Average Math Proficiency')
            plt.xticks(x, labels)

            plt.tight_layout()
            plt.show()
        else:
            print("Insufficient data available to plot both Reading and Math proficiency.")

    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
plot_proficiency_levels(df)
```

```python
# Khả năng đọc thành thạo theo quốc gia
fig1 = px.choropleth(df, locations='location', locationmode='country names',
                    color='Grade_2_3_Proficiency_Reading', color_continuous_scale='Viridis',range_color=[0, 100],
                    title='Proficiency in Reading by Country')
fig1.show()

# Trình độ Toán theo quốc gia

fig2 = px.choropleth(df, locations='location', locationmode='country names',
                    color='Grade_2_3_Proficiency_Math',color_continuous_scale='Viridis', range_color=[0, 100],
                    title='Proficiency in Math by Country')
fig2.show()
```

# ***Tỉ lệ hoàn thành các bậc giáo dục của nam và nữ***

```python
def plot_completion_rates(df):
    try:
        # Cột biểu thị tỷ lệ hoàn thành theo trình độ học vấn và giới tính
        completion_columns = [
            'Completion_Rate_Primary_Male', 'Completion_Rate_Primary_Female',
            'Completion_Rate_Lower_Secondary_Male', 'Completion_Rate_Lower_Secondary_Female',
            'Completion_Rate_Upper_Secondary_Male', 'Completion_Rate_Upper_Secondary_Female'
        ]

        # Trích xuất các cột tồn tại trong tập dữ liệu
        available_columns = [col for col in completion_columns if col in df.columns]

        # Tính tỷ lệ hoàn thành trung bình theo giới tính cho từng cấp học
        average_values = [df[col].mean() for col in available_columns]

        if len(average_values) >= 2:  # Kiểm tra xem có giá trị nào cần vẽ cho cả hai giới tính không
            labels = ['Primary', 'Lower Secondary', 'Upper Secondary']
            male_completion = average_values[::2]  # Trích xuất tỷ lệ hoàn thành của nam giới
            female_completion = average_values[1::2]  # Trích xuất tỷ lệ hoàn thành của nữ

            x = np.arange(len(labels))
            width = 0.35  # chiều rộng của thanh

            plt.figure(figsize=(10, 6))

            # Biểu đồ thanh về Tỷ lệ hoàn thành theo giới tính
            plt.bar(x - width/2, male_completion, width, label='Male', color='skyblue')
            plt.bar(x + width/2, female_completion, width, label='Female', color='salmon')

            plt.xlabel('Education Levels')
            plt.ylabel('Completion Rate')
            plt.title('Completion Rates Across Education Levels and Gender')
            plt.xticks(x, labels)
            plt.legend()
            plt.tight_layout()
            plt.show()
        else:
            print("Insufficient data available to plot completion rates by gender.")

    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
plot_completion_rates(df)
```

# ***So sánh tỷ lệ sinh với tỷ lệ tuyển sinh vào đại học ***

```python
def plot_birth_vs_enrollment(df):
    try:
        # Selecting columns for birth rate and gross enrollment in primary/tertiary education
        columns = ['Birth_Rate', 'Gross_Primary_Education_Enrollment', 'Gross_Tertiary_Education_Enrollment']

        # Filtering data for rows with non-null values in the selected columns
        filtered_data = df.dropna(subset=columns)

        # Extracting data for plotting
        birth_rates = filtered_data['Birth_Rate']
        primary_enrollment = filtered_data['Gross_Primary_Education_Enrollment']
        tertiary_enrollment = filtered_data['Gross_Tertiary_Education_Enrollment']

        plt.figure(figsize=(10, 8))

        # Scatter plot for birth rates vs. gross enrollment in primary education
        plt.scatter(birth_rates, primary_enrollment, s=50, alpha=0.6, label='TS tiểu học')

        # Bubble chart for birth rates vs. gross enrollment in tertiary education
        plt.scatter(birth_rates, tertiary_enrollment, s=50*np.sqrt(tertiary_enrollment),
                    alpha=0.6, label='TS đại học')

        plt.xlabel('Tỷ lệ sinh')
        plt.ylabel('Tỷ lệ tuyển sinh')
        plt.title('Tỷ lệ sinh vs. Tỷ lệ tuyển sinh')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
plot_birth_vs_enrollment(df)
```

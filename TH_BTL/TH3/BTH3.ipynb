```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/KPDL/TH3/input/Full_Mark_2020.csv")

df.columns = ['No', 'Code', 'Diali', 'GDCD', 'Hoahoc', 'KHTN', 'KHXH', 'LichSu', 'Ngoaingu', 'Nguvan', 'Sinhhoc', 'Toan', 'Vatli', 'city']

df.info()

df
```

```python
df["Code"] = pd.to_numeric(df['Code'],errors='coerce')
df["Vatli"] = pd.to_numeric(df['Vatli'],errors='coerce')
df["city"] = pd.to_numeric(df['city'],errors='coerce')
df.info()
```

```python
# Kiểm tra dữ liệu trùng lặp
duplicate_rows = df[df.duplicated(subset="Code")]
print('Số dòng dữ liệu trùng lặp:', len(duplicate_rows))
# df.drop_duplicates(inplace=True) # Xóa dữ liệu trùng lặp
```

```python
# Sao chép df ra my_df và tiến hành đánh số lại cho cột Code
my_df = df
my_df["Code"] = range(1, len(df) + 1)

my_df
```

```python
# Đếm số lượng thí sinh không thi cả hai môn KHTN và KHXH
count_no_exam = len(my_df[(my_df['KHTN'].isnull()) & (my_df['KHXH'].isnull())])
# Đếm số lượng thí sinh thi cả hai môn KHTN và KHXH
count_both_exam = len(my_df[(my_df['KHTN'].notnull()) & (my_df['KHXH'].notnull())])

# In kết quả
print("Số lượng thí sinh không thi cả hai môn: ", count_no_exam)
print("Số lượng thí sinh dự thi cả hai môn: ", count_both_exam)
```

```python
# Tách DataFrame my_df thành my_df_khtn và my_df_khxh
my_df_khtn = my_df[my_df['KHTN'].notnull()]
my_df_khxh = my_df[my_df['KHXH'].notnull()]
my_df_khtn.info()
my_df_khxh.info()
```

```python
# Lọc dữ liệu điểm thi của thí sinh tại Tp.HCM
df_khtn_hcm = my_df_khtn[my_df_khtn["city"] == 2]
df_khtn_hcm.to_csv("/content/drive/MyDrive/KPDL/TH3/output/df_khtn_hcm.csv")

# Lọc dữ liệu điểm thi của thí sinh tại Tp.HCM
df_khxh_hcm = my_df_khxh[my_df_khxh["city"] == 2]
df_khxh_hcm.to_csv("/content/drive/MyDrive/KPDL/TH3/output/df_khxh_hcm.csv")
```

```python
# Lọc ra các cột cần tính toán
my_df_khtn = my_df_khtn[["Toan", "Nguvan", "Ngoaingu", "Vatli", "Hoahoc", "Sinhhoc"]]
my_df_khtn.info()
my_df_khxh = my_df_khxh[["Toan", "Nguvan", "Ngoaingu", "LichSu", "Diali", "GDCD"]]
my_df_khxh.info()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Tiền xử lý dữ liệu
my_df_khtn.fillna(my_df_khtn.mean(), inplace=True)
```

```python
# Tìm số cụm tối ưu sử dụng phương pháp Elbow
max_clusters = 10  # Số cụm tối đa muốn xem
sse = []
for k in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(my_df_khtn)
    sse.append(kmeans.inertia_)

# Vẽ biểu đồ Elbow
plt.plot(range(1, max_clusters + 1), sse, marker='o')
plt.xlabel('Số lượng cụm')
plt.ylabel('SSE')
plt.title('Phương pháp Elbow')
plt.show()
```

```python
# Số cụm tối ưu dựa trên biểu đồ Elbow
num_clusters = 5

# Áp dụng thuật toán K-means
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(my_df_khtn)

# Gán nhãn cụm cho dữ liệu
my_df_khtn['Cluster'] = kmeans.labels_

# Vẽ biểu đồ phân cụm
plt.scatter(my_df_khtn['Cluster'], my_df_khtn['Toan'], c=my_df_khtn['Cluster'], cmap='viridis')
plt.xlabel('Cluster')
plt.ylabel('Toan')
plt.title('Phân cụm K-means')
plt.show()
```

```python
my_df_khtn[["Toan", "Nguvan"]].round(2)
```

```python
import numpy as np
def kmeans_display(X, label):
    K = np.amax(label) + 1
    X0 = X[label == 0, :]
    X1 = X[label == 1, :]
    X2 = X[label == 2, :]

    plt.plot(X0[:, 0], X0[:, 1], 'b^', markersize = 4, alpha = .8)
    plt.plot(X1[:, 0], X1[:, 1], 'go', markersize = 4, alpha = .8)
    plt.plot(X2[:, 0], X2[:, 1], 'rs', markersize = 4, alpha = .8)

    plt.xlabel("Toan")
    plt.ylabel("Van")
    plt.axis('equal')
    plt.plot()
    plt.show()
# Số cụm tối ưu dựa trên biểu đồ Elbow
num_clusters = 5

# Áp dụng thuật toán K-means
kmeans = KMeans(n_clusters=num_clusters, random_state=0).fit(my_df_khtn[["Toan", "Nguvan"]])

print("Center found by scikit-learn:\n", kmeans.cluster_centers_)

pred_label = kmeans.predict(my_df_khtn[["Toan", "Nguvan"]])

kmeans_display(my_df_khtn[["Toan", "Nguvan"]].values, pred_label)
```

```python
import pandas as pd
import csv

# Specify the file path
file_path = "/content/drive/MyDrive/KPDL/TH3/input/car.csv"
car_df = pd.read_csv(file_path, encoding='utf-8', delimiter='|', on_bad_lines='skip')

car_df.info()

# car_df#.to_csv("/content/drive/MyDrive/Colab Notebooks/KPDL/BTH03/car_new.csv")
```

```python
car_df = car_df.drop(columns=["km_1", "imp_exp_1", "poster_name", "poster_add", "poster_tel", "title"])

# Chuẩn hóa cột km
car_df['km'] = car_df["km"].str.replace(' Km', '')
car_df['km'] = car_df["km"].str.replace(',', '').astype(int)

# Chuẩn hóa cột door_num
car_df['door_num'] = car_df["door_num"].str.replace(' cửa', '').astype(int)

# Chuẩn hóa cột seat_num
car_df['seat_num'] = car_df["seat_num"].str.replace(' chỗ', '').astype(int)

car_df["price"] = car_df["price"].str.replace('Tỷ', '')
car_df["price"] = car_df["price"].str.replace('Triệu', '')
car_df["price"] = car_df["price"].str.replace(' ', '').astype(int)


car_df
```

```python
# Tìm số cụm tối ưu sử dụng phương pháp Elbow
max_clusters = 10  # Số cụm tối đa muốn xem
sse = []
for k in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(car_df[["price"]])
    sse.append(kmeans.inertia_)

# Vẽ biểu đồ Elbow
plt.plot(range(1, max_clusters + 1), sse, marker='o')
plt.xlabel('Số lượng cụm')
plt.ylabel('SSE')
plt.title('Phương pháp Elbow')
plt.show()
```

```python
# Số cụm tối ưu dựa trên biểu đồ Elbow
num_clusters = 4

# Áp dụng thuật toán K-means
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
kmeans.fit(car_df[["price"]])

# Gán nhãn cụm cho dữ liệu
car_df['Cluster'] = kmeans.labels_

# Vẽ biểu đồ phân cụm
plt.scatter(car_df['seat_num'], car_df['price'], c=car_df['Cluster'], cmap='viridis')
plt.xlabel('seat_num')
plt.ylabel('price')
plt.title('Phân cụm K-means')
plt.show()
```

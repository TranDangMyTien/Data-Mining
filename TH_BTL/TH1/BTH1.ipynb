# Load dữ liệu sinh viên từ file xlsx

```python
# Import thư viện pandas:
import pandas as pd

# Đọc dữ liệu từ 4 file Excel và lưu vào các DataFrame:
# Đọc dữ liệu từ file DSSV1
df1 = pd.read_excel('/content/drive/MyDrive/KPDL/TH1/input/DSSV1.xlsx')

# Đọc dữ liệu từ file DSSV2
df2 = pd.read_excel('/content/drive/MyDrive/KPDL/TH1/input/DSSV2.xlsx')

# Đọc dữ liệu từ file DSSV3
df3 = pd.read_excel('/content/drive/MyDrive/KPDL/TH1/input/DSSV3.xlsx', header=0, names=['STT', 'MSSV', 'Ho', 'Ten', 'NgayThang'])

# Đọc dữ liệu từ file DSSV4
df4 = pd.read_excel('/content/drive/MyDrive/KPDL/TH1/input/DSSV4.xlsx')
```

# Chuẩn hóa dữ liệu của từng DataFrame:

```python
# Chuẩn hóa dữ liệu của DataFrame df1
df1 = df1.rename(columns={'Ho5': 'Ho', 'Ngay': 'Ngay', 'Thang': 'Thang', 'Nam': 'Nam'})
# Điền giá trị 0
df1 = df1.fillna(0)

# Chuyển đổi dữ liệu các cột Ngay, Thang, Nam thành kiểu int
df1['Ngay'] = df1['Ngay'].astype(int)
df1['Thang'] = df1['Thang'].astype(int)
df1['Nam'] = df1['Nam'].astype(int)

print(df1)
```

```python
# Tách cột ngày tháng năm thành 3 cột riêng trong df2
df2[['Nam', 'Thang', 'Ngay']] = df2['yyyy-mm-dd'].str.split('-', expand=True)
# Điền giá trị 0
df2 = df2.fillna(0)

# Chuyển đổi kiểu dữ liệu của các cột Ngay, Thang, Nam thành int
df2['Ngay'] = df2['Ngay'].astype(int)
df2['Thang'] = df2['Thang'].astype(int)
df2['Nam'] = df2['Nam'].astype(int)

# Xóa cột ngày tháng năm cũ
df2 = df2.drop('yyyy-mm-dd', axis=1)

# Đặt lại tên các cột
df2 = df2.rename(columns={'Họ Lót': 'Ho', 'Tên': 'Ten'})

df2
```

```python
# Tách cột ngày tháng năm thành 3 cột riêng trong df3
df3[['Ngay', 'Thang', 'Nam']] = df3['NgayThang'].str.split('/', expand=True)
# Điền giá trị 0
df3 = df3.fillna(0)

# Chuyển đổi kiểu dữ liệu của các cột Ngay, Thang, Nam thành int
df3['Ngay'] = df3['Ngay'].astype(int)
df3['Thang'] = df3['Thang'].astype(int)
df3['Nam'] = df3['Nam'].astype(int)

# Xóa cột ngày tháng năm cũ
df3 = df3.drop('NgayThang', axis=1)

df3
```

```python
# Tách cột ngày tháng năm thành 3 cột riêng trong df4
df4[['Ngay', 'Thang', 'Nam']] = df4['dd-mm-yyyy'].str.split('-', expand=True)
# Điền giá trị 0
df4 = df4.fillna(0)

# Chuyển đổi kiểu dữ liệu của các cột Ngay, Thang, Nam thành int
df4['Ngay'] = df4['Ngay'].astype(int)
df4['Thang'] = df4['Thang'].astype(int)
df4['Nam'] = df4['Nam'].astype(int)

# Xóa cột ngày tháng năm cũ
df4 = df4.drop('dd-mm-yyyy', axis=1)

# Đặt lại tên các cột
df4 = df4.rename(columns={'Họ Lót': 'Ho', 'Tên': 'Ten'})

df4
```

```python
# Tạo DataFrame tổng hợp từ các DataFrame đã chuẩn hóa:
df_combined = pd.concat([df1, df2, df3, df4], ignore_index=True)

# Chuẩn hóa kiểu dữ liệu
df_combined['MSSV'] = df_combined['MSSV'].astype(int)
df_combined['Ho'] = df_combined['Ho'].astype(str)
df_combined['Ten'] = df_combined['Ten'].astype(str)

df_combined
```

```python
# Thay thế các chữ sau trong các cột tương ứng của DataFrame df_combined
df_combined = df_combined.replace({'Ngũyên': 'Nguyễn', 'Hòang': 'Hoàng', 'Hùynh': 'Huỳnh', 'ơưng': 'ương', 'Túân': 'Tuấn'}, regex=True)
df_combined
```

```python
# Xóa cột STT
df_combined = df_combined.drop('STT', axis=1)

# Tạo DataFrame mới lưu trữ dòng dữ liệu không hợp lệ ban đầu
df_invalid = df_combined.copy()

# Hàm kiểm tra tính hợp lệ của ngày tháng
def is_valid_date(row):
    try:
        day = int(row['Ngay'])
        month = int(row['Thang'])
        year = int(row['Nam'])
        pd.Timestamp(year=year, month=month, day=day)
        return True
    except:
        return False

# Hàm kiểm tra tính hợp lệ của cột MSSV, Ho, Ten
def is_valid_data(row):
    if row['MSSV'] == '0' or row['Ho'] == '0' or row['Ten'] == '0':
        return False
    return True

# Áp dụng hàm kiểm tra tính hợp lệ cho từng dòng dữ liệu
mask = df_combined.apply(lambda row: is_valid_date(row) and is_valid_data(row), axis=1)

# Chia dữ liệu thành df_valid và df_invalid dựa trên kết quả kiểm tra
df_valid = df_combined[mask]
df_invalid = df_combined[~mask]

# Chuyển đổi kiểu dữ liệu cho cột Ngay, Thang, Nam của df_valid thành int
df_valid['Ngay'] = df_valid['Ngay'].astype(int)
df_valid['Thang'] = df_valid['Thang'].astype(int)
df_valid['Nam'] = df_valid['Nam'].astype(int)
```

```python
# Lưu DataFrame df_valid thành file Excel
df_valid.to_excel('df_valid.xlsx', index=False)

# Lưu DataFrame df_invalid thành file Excel
df_invalid.to_excel('/content/drive/MyDrive/Colab Notebooks/KPDL/CAC BUOI THUC HANH/BTH01/df_invalid.xlsx', index=False)

df_valid['Nam'].value_counts().plot(kind='bar')
```
